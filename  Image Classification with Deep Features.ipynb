```python
image_test = graphlab.SFrame('image_test_data/')
```

# Triaing a classifer on raw images pixels

```python
raw_pixel_model = graphlab.logistic_classifier.create(image_train,target='label',features=['image_array'])
```

# Improving the model use deep features

```python
len(image_train)
```

```python
deep_learning_model = graphlab.load_model('imagenet_model')#alexnet model pre trained on graohlab
#for transfer learning
image_train['deep_features'] = deep_learning_model.extract_features(image_train)
```

```python
image_train.head()
```

# Given the deep features, Training a classifer

```python
deep_features_model = graphlab.logistic_classifier.create(image_train,features=['deep_features'],target='label')
```

# Applying the deep features model to the first few datasets

```python
image_test[0:3]['image'].show()
```

```python
deep_features_model.predict(image_test[0:4])
```

# Computing the test data accuracy of deep features model

```python
deep_features_model.evaluate(image_test)#78% accuracy way better than raw pixel model which had 46.8% accuracy
```

# Traing a nearest neighbour model for reteriving Images using deep features

# KNN MODEL

```python
knn_model = graphlab.nearest_neighbors.create(image_train,features=['deep_features'],label='id')
```

# Using image retrieval model with deep features to fing similar images

```python
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)#to find nearest neighbours
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neighbours =get_images_from_ids(knn_model.query(cat))
```
